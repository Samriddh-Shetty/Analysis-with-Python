#### WE CAN FIND PROBABLY FIND THE TOTAL MORE ABOUT THE SUMMATION OF returns after 1,3 and 5 years  using compound interest formula but decided not to do that because its 
#### heavily dependent on the NAV price at the time of purchase and don't have either of those data in our columns

```python

principal = float(input("ENTER THE TOTAL AMOUNT TO BE INVESTED"))
annual_rate = float(input("ENTER THE DESIRED INTEREST RATE"))
time_years = int(input("ENTER THE NO OF YEARS"))

if principal>=0 and time_years>0:
    if time_years==1:
        print(df[df["returns_1yr"]==annual_rate])
    elif time_years==3:
        print(df[df["returns_3yr"]==annual_rate])
    elif time_years==5:
        print(df[df["returns_5yr"]==annual_rate])
        

    amount = principal * (1 + annual_rate / 100) ** time_years
    print("The total amount after {} years of compound interest is: {:.2f}".format(time_years, amount))
else:
    print("Invalid input. Principal and time_years must be non-negative.")
```

#### I CREATED A SIMPLE FILTERED FUNCTION USING SIMPLE COMPOUND INTEREST WHICH WILL TELL US HOW MUCH RETURN A PARTICULAR MUTUAL FUNDS WILL GIVE AFTER A SPECIFIC INTERVAL OF TIME.
